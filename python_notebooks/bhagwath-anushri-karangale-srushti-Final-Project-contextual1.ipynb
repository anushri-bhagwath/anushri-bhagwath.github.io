```python
#importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
```

For this contexual visual we are using New York City Community Health Survey dataset and a detailed information about the dataset can be found using the link https://catalog.data.gov/dataset/dohmh-community-health-survey-2010-2016. We have used this dataset in a csv format and its link is https://data.cityofnewyork.us/api/views/csut-3wpr/rows.csv?accessType=DOWNLOAD .This dataset contains the data regarding the health of New Yorkers, including neighborhood, borough, and citywide estimates on a broad range of chronic diseases and behavioral risk factors. We have used this dataset to visualize the percentage of patients without proper medical care over the years between 2010 and 2015 in New York City.

```python
# Reading the data
NYC_survey_data = pd.read_csv('https://data.cityofnewyork.us/api/views/csut-3wpr/rows.csv?accessType=DOWNLOAD').dropna()
```

```python
# displaying the few rows of the dataset
NYC_survey_data.head()
```

```python
# check the dataset and its datatypes
NYC_survey_data.info()
```

Prevelance means the proportion of persons who have a condition at or during a particular time period. Hence we wanted to just filter out the dataset and obtain the prevelance values for the years between 2010 and 2015. 

```python
NYC_cause_prevalent = NYC_survey_data.query("Prevelance in ('Prevalence 2011', 'Prevalence 2012','Prevalence 2013','Prevalence 2014')")
```

```python
# display the filtered data
NYC_cause_prevalent
```

We decided to plot a graph visualizing the percentage of patient in New York City who didn't get proper medical care for the years between 2010 and 2015.

```python
# using altair to plot the bar chart
bar_chart =  alt.Chart(NYC_cause_prevalent).mark_bar().encode(
  alt.X('Year:N'),  # N is for nominal
  alt.Y('Did not get needed medical care:Q', axis=alt.Axis(grid=False), title ='Percentage of patients without proper medical care'), # Q is for quantitative
  tooltip=(['Did not get needed medical care:Q'])
  ).properties(
    title='Percentage of patients without proper medical care over the years'
)
```

```python
# display the bar chart
bar_chart = bar_chart.properties(width=500)
bar_chart
```

#### Plot choice explanation:

We chose the bar plot for this visual because we felt that it will help us see the change in the percentage of patients without proper medical care over the four years. The choice of color was not consciously made.

While choosing the contextual visualizations, we had to look for datasets that would help us learn more and relate to our primary dataset. It had various fields which provides us added information such as percentage of people with no access to medical care, access to health insurance, their lifestyle choices such as diet, smoking habits etc. We considered this particular dataset since these factors could provide us extra information to how these lifestyle choices and external factors are also a contributor to the cause of death in NYC. We plotted a simple bar graph for this visualization that shows us the percentage of people without medical care from the years 2011 to 2014.
